#### Functions for Data Extraction from HTML
Below are the five functions that were used to parse the raw HTML data from its intial state into usable data. The five separate functions return a list of data for each job posting, returning an `N/A` value for any postings without data (of which there were many in the salary field, as expected/pre-warned).

```python
def get_remote(result):
    '''
    Function to retrieve remote working Yes/Temp/No from HTML
    '''
    page     = BeautifulSoup(result,"html.parser")
    jobcards = page.find_all("div",class_="jobsearch-SerpJobCard")
    
    return [card.find("span",class_="remote").text.strip()
            if card.find("span",class_="remote") != None
            else "Not Remote"
            for card in jobcards]

def get_company(result):
    '''
    Function to retrieve company names from HTML
    '''
    page     = BeautifulSoup(result,"html.parser")
    jobcards = page.find_all("div",class_="jobsearch-SerpJobCard")
    
    return [card.find("span",class_="company").text.strip()
            if card.find("span",class_="company") != None
            else np.nan
            for card in jobcards]

def get_title(result):
    '''
    Function to retrieve job titles from HTML
    '''
    page     = BeautifulSoup(result,"html.parser")
    jobcards = page.find_all("div",class_="jobsearch-SerpJobCard")

    return [card.find("a",class_="jobtitle turnstileLink").text.strip()
            if card.find("a",class_="jobtitle turnstileLink") != None
            else np.nan
            for card in jobcards]

def get_location(result):
    '''
    Function to retrieve locations from HTML
    '''
    page     = BeautifulSoup(result,"html.parser")
    jobcards = page.find_all("div",class_="jobsearch-SerpJobCard")

    return [card.find("div",class_="location").text.strip()
            if   card.find("div",class_="location") != None
            else card.find("span",class_="location").text.strip()
            if   card.find("span",class_="location") != None
            else np.nan
            for  card in jobcards]

def get_salary(result):
    '''
    Function to retrieve salaries from HTML
    '''
    page     = BeautifulSoup(result,"html.parser")
    jobcards = page.find_all("div",class_="jobsearch-SerpJobCard")

    return [card.find("span",class_="salaryText").text.strip()
            if card.find("span",class_="salaryText") != None
            else np.nan
            for card in jobcards]
```

We now need to convert the strings in the salary field into number formats, taking the mid-point of any ranges given. The `.split()` function will be able to separate these strings, but we need to be sure that there are only two formats of salary given (`£X a year` and `£Y - £Z a year`) by using:
```python
ann_slry['salary'].apply(lambda x: len(x.split())).value_counts()
```
which yields:
```
5    1635
3     637
Name: salary, dtype: int64
```
So, as we can see from above, only lists of length 3 (`£X a year` format) and 5 (`£Y - £Z a year` format) are generated, therefore we can then write the below function to then `.apply()` onto the DataFrame to arrive at our desired result.
```python
def salary_convert(salary):
    '''
    Function to convert salary strings to floats
    '''
    splitter = salary.split()
    if len(splitter) == 3: # For single salaries given
        return float(splitter[0].replace("£","").replace(",",""))
    else:                  # For salary ranges given
        lo =   float(splitter[0].replace("£","").replace(",",""))
        hi =   float(splitter[2].replace("£","").replace(",",""))
        return (lo+hi)/2.0
```

#### Reduction in Unique Locations
Another issue is the very large number of unique locations that our code has pulled into the DataFrame, which will make dummification and classification difficult, with a large number of very low-population classes.

One of the reasons for this is that some locations suffix postcodes on to their locations, resulting in duplicates not picked up by the previous data munging processes. Writing another simple function to apply onto the DataFrame helps to truncate this list by a meaningful amount. For example:
```python
ann_slry['location'].unique()[:5]
```
```
>>> array(['London', 'London NW1', 'Ilford', 'London E20 1JN',
       'Birmingham B3 2PW'], dtype=object)
```
is enough to show us the essence of the problem at hand, with multiple London locations being perceived as unique. The following function manages to remove this problem for us by once again using `.split()` and then RegEx to filter out any strings containing numbers.
```python
def drop_pstcde(location):
    '''
    Function to remove post-code suffixes to locations
    '''
    new_word = ""
    counter  = 0
    for word in location.split():
        if re.findall("\d",word) == []:
            counter  += 1
            if counter > 1:
                new_word += " "+word
            else:
                new_word += word
    return new_word
```
We can check how effective this filter will be by a simple comparison pre- and post-application.

#### Establishing High / Low Salaries
We have divided our salary data into two classes, with a salary being classed as "High" if it is above the median and "Low" otherwise. We can do this utilising the `.apply` function once again, using a simply lambda boolean function.

````python
salary_median = scrape['salary'].median()

scrape['hi_lo'] = scrape['salary'].apply(lambda x: "High" if x > salary_median else "Low")
````

We can compute our high/low salary classification baseline by looking at the value counts in each class - as we have separated the data by the median this is purely a formality as we know the classes should be equally balanced.

````python
scrape['hi_lo'].value_counts(normalize=True)
````

Now that we have divided our dataset into these two classes, we can do some quick exploration of any relationships that might be of interest. Given the large number of locations, we will look at the top-20 most populated locations for ease.

````python
# Low salary locations
low_sal = pd.DataFrame()
low_sal["Low"] = scrape[scrape["hi_lo"]=="Low"]['location'].value_counts()
# High salary locations
hi_sal = pd.DataFrame()
hi_sal["High"] = scrape[scrape["hi_lo"]=="High"]['location'].value_counts()
# All locations
all_sal = pd.DataFrame()
all_sal["Total"] = scrape['location'].value_counts()
# Replace NaN values with 0 as any class with 0 population will have been filled as a NaN when joining
compare_loc = pd.concat([low_sal,hi_sal,all_sal], axis=1,join="outer").fillna(0)
compare_loc["Low %"]  = round(100*compare_loc["Low"]/compare_loc["Total"])
compare_loc["High %"] = round(100*compare_loc["High"]/compare_loc["Total"])
# Sort the table in descending order of the population of observations for a given location
compare_loc.sort_values(by="Total",ascending=False,inplace=True)
compare_loc.head(20)
````

````python
# Representing this visually for the top-20
fig,ax = plt.subplots(figsize=(6,6))

sns.heatmap(compare_loc[["Low %","High %"]].head(20),cmap="Greens",annot=True)
plt.show()
````

From this cursory analysis, we can certainly see relationships above and below the naive 50% baseline assumption, which provides support that there could indeed be a relationship we can model.

Next, we can iterate through a list of simple classification models, to see which model could be the best fit for our dataset.

````python
regsvc = SVC()
linsvc = LinearSVC()
logreg = LogisticRegression(max_iter=10000)
d_tree = DecisionTreeClassifier()
r_tree = RandomForestClassifier(n_estimators=1000,max_depth=3)
e_tree = ExtraTreesClassifier()
knn    = KNeighborsClassifier(n_neighbors=10)

classifiers = [ regsvc,  linsvc,  logreg,  d_tree,  r_tree,  e_tree,  knn]
class_label = ["regsvc","linsvc","logreg","d_tree","r_tree","e_tree","knn"]

kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)

y = scrape['hi_lo']
X = scrape['location']

X_dum = pd.get_dummies(X)

Xtr,Xte,ytr,yte = train_test_split(X_dum,y,stratify=y,test_size=0.2,random_state=1)

trains = []
tests  = []
cvs    = []
delay  = []

for model in classifiers:
    start_time = time.time()
    model.fit(Xtr,ytr)
    tr = model.score(Xtr,ytr)
    te = model.score(Xte,yte)
    cv = cross_val_score(model,X_dum,y,cv=kf).mean()
    ti = time.time()-start_time
    trains.append(tr)
    tests.append(te)
    cvs.append(cv)
    delay.append(ti)
    print(f'{model}')
    print('-'*32)
    print(f'Train Score:  {tr}')
    print(f'Test Score:   {te}')
    print(f'CV Score:     {cv}')
    print(f'Compute time: {round(ti,2)} seconds')
    print('-'*32)
    print()
    
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(ncols=2,nrows=2,figsize=(12,12))
ax1.plot(class_label,trains,marker="o")
ax1.set_title("\nTraining Scores\n")
ax2.plot(class_label,tests,marker="o")
ax2.set_title("\nTest Scores\n")
ax3.plot(class_label,cvs,marker="o")
ax3.set_title("\nCV Scores\n")
ax4.bar(class_label,delay)
ax4.set_title("\nCompute Time\n")
plt.show()
````

From the above results, we can see that there is not a huge amount of difference between the cross-validated scores, aside from K-Neighbours which was a fairly notable underperformer versus the other models, although still better than the naive baseline of 50%. `SVC`, `LinearSVC` and `LogisticRegression` also performed strongly on unseen test data, whereas the decision tree methods exhibited some potential early signs of over-fitting and perhaps not generalising well - bar the `RandomForestClassifier` on the test data, but it still scored poorly on cross validation - although in reality the absolute difference in accuracy scores was small (K-Neighbours excluded).

Given the comparable scores, both on a stratified K-fold cross-validation basis and on unseen test data, and impressive computation speed we will likely try to further improve upon our `LogisticRegression` score by attempting to optimise our parameters, but first we can check further on how well each of the models performed on a more granular basis by using the classification report and ROC/precision-recall curves on the whole dataset (but having only been fitted using the training data).

If we revisit the histogram from earlier, it is perhaps easier to see why the confusion around the low-salary class might be the case.

````python
# From above, reprinted with sigma levels from the median

salary_med = scrape['salary'].median()
salary_std = scrape['salary'].std()

fig,ax = plt.subplots(figsize=(10,6))
bins = range(0,162500,2500)
plt.hist(scrape['salary'],bins=bins,edgecolor='black',label="£2,500 Salary Bins")
plt.ylim(0,200)
ax.set_ylabel("Frequency Density")
ax.set_xlabel("Salary")
for sigma in range(-2,9,1):
    plt.axvline(salary_med+(sigma*salary_std),linestyle="--",color="blue")
    if sigma < 0:
        plt.annotate(f"{sigma}\u03C3",(salary_med+(sigma*salary_std)+1250,168))
    elif sigma == 0:
        plt.annotate("Median",(salary_med+1000,168))
    else:
        plt.annotate(f"+{sigma}\u03C3",(salary_med+(sigma*salary_std)+1250,168))
plt.legend()
plt.show()
````

We can clearly see that the distribution is postiively skewed, with the "High" salary classification data spread over a far larger area than the "Low" salary classification data; the entirety of the "Low" salary data is distributed from the median over a space slightly smaller than 2 standard deviations, whereas the "High" salary data is distributed across a region exceeding 7 standard deviations; admittedly the bulk of the data is within the 2 standard deviation range, but there are still a meaningful number of observations above this point.

#### Changing Thresholds to Meet Specific Purposes
In our model to prevent giving false negatives (telling someone a role would be "High" salary but in fact being "Low" salary, in the context of our model) while tolerating false positives, you would move the threshold for predicting the "Low" salary class down from 50% to a lower value, making it more likely to predict a role as "Low" salary but also less likely to incorrectly predict a role as "High" salary.

For example, we would use the model to predict the probabilities (rather than the values themselves) and can manually change the threshold levels.

````python
# Let us lower the tolerance level for being predicted "Low" salary to 35%, for e.g.
predictions = pd.DataFrame(logreg.predict_proba(X_dum),columns=['prob_hi','prob_lo'])
predictions['y_true'] = y
predictions['y_pred'] = predictions['prob_lo'].apply(lambda x: "Low" if x > 0.35 else "High")
# Add columns with conditionals to sum true and false positives and negatives
predictions['fp'] = ((predictions["y_true"]== "High") & (predictions["y_pred"]=="Low"))
predictions['fn'] = ((predictions["y_true"]== "Low")  & (predictions["y_pred"]=="High"))
predictions['tp'] = ((predictions["y_true"]== "Low")  & (predictions["y_pred"]=="Low"))
predictions['tn'] = ((predictions["y_true"]== "High") & (predictions["y_pred"]=="High"))
fp = predictions['fp'].sum()
fn = predictions['fn'].sum()
tp = predictions['tp'].sum()
tn = predictions['tn'].sum()
# Assemble the new confusion matrix
conf_matrix = np.array([tn,fp,
                        fn,tp]).reshape(2,2)
# Plot the confusion matrix
fig,ax = plt.subplots(figsize=(8,6))
sns.heatmap(pd.DataFrame(conf_matrix),cmap="Blues",annot=True,)
plt.show()
````

Comparing the confusion matrix above to the one below (`1 = "Low"` salary, `0 = "High"` salary) you can see by moving the threshold in this manner we have greatly reduced the number of false negatives (i.e. telling someone a job will be "High" salary when in fact it will be "Low" salary) but at the cost of increasing the false positives (i.e. the exact inverse) as well. Dependent on the requirement from the model, this may not necessarily be a problem - but it is all down to context.

````python
# Calculate some other measures of this 35% tolerance model
accuracy  = (tp+tn)/(tp+tn+fp+fn)
precision = tp/(tp+fp)
recall    = tp/(tp+fn)
print(f"Accuracy  @ 35% Tolerance: {round(accuracy,4)}")
print(f"Precision @ 35% Tolerance: {round(precision,4)}")
print(f"Recall    @ 35% Tolerance: {round(recall,4)}")
````

A desire for high precision in this context would be when you are focused on ensuring you are correctly identifying a class, minimising false positives. In the frame of my models, this would mean ensuring not to indicate a role would be "Low" salary when it was in fact "High" salary, but the framing is arbitrary and the inverse would also be true, depending on your perspective.

A desire for a high recall score in the context of this modelling would be a desire to classify one or the other class well, minimising false negatives. Once again, in the frame of my current model, this would be to ensure all the jobs clssified as "Low" salary were classified as such, and none were accidentally classified as "High" - once again, the inverse is also true depending on how you frame your models initially.

### Summary

##### Abstract

We set out to investigate whether there was a relationship between the geographical location of a job posting on a market-leading jobs board and its advertised salary, where salary information was available (a large proportion of salaries are not disclosed on such job postings).

Initially we tried to gauge the relationships between locations and salaries and found there to be strong evidence of meaningful and potentially predictable relationships amongst higher paying roles, but correctly predicting lower end salaries was much more difficult with the information we had available.

Adaptations were made to our intial model to include other variables in our search, where we carried keyword searches relating to seniority of roles, including ones that specified roles were junior as we believed this would be a good way to provide more insight and increase our ability to delineate lower paying roles from higher paying ones.

This did indeed result in a better overall result, although the improvements were only modest. Given the methodology followed during this investigation, we can say that predictions we will be currently able to make will be an improvement on randomised guessing - but further refinements and extra information - particularly around the low salary roles - would be a key avenue for improving our ability to correctly classify job salaries going forward.

##### Methodology

We utilised web scraping technology - whereby a piece of code routinely downloads information from publicly available websites - to aggregate a large number of individual job postings (ensuring duplicates had been removed) and categorised each job's location, remote working status, salary, job title and company name.

Once this data had been collected and cleaned (as information is often duplicated or missing, especially salaries as mentioned earlier), we analysed how the range of salaries were distributed versus the number of job postings within each given range of salaries. What was initially apparent was the roughly bell-curve shaped distribution with a long tail towards the higher-end of salaries.

We modelled these relationships by analysing which locations had the most job postings in the high and low salary bands (categorised as being above or below the median of our sample set in this case) and then trying to assess the probability of a job posting being either in the high or low salary classification, based on its location and the salaries of other such jobs in that location, and therefore make predictions based on the outcome with the highest likelihood, according to what was suggested by the dataset we had available from our web scraping exercise.

The long and thin tail over the higher range of salaries meant that high salary data was less densely packed, and therefore more easily demarcated and subsequently classified. However, the bulk of the data being concentrated within a fairly narrow range of salaries lower in the overall range made classifying the remaining mostly lower salary job postings more difficult, even initially resulting in one of the models we tested only slightly outperforming randomised approximation.

We ultimately needed more contextualised and granular information about these roles to be able to improve on our delineation of whether or not they fell into either the high or low salary classification.

Introducing further categories relating to whether the job title had certain keywords in (junior, graduate, senior, manager) provided some of this necessary clarity, and unsurprisingly the two new categories we introduced that you might logically expect to be at the lower end of the salary range (junior and graduate) did indeed hold a strong importance within our models. This reassured our initial base assumption that further context to the lower end of salaries was likely where the most quick and easy gains were to be made in regards to improving accurate predictive power by offering additional methods of classifying a job posting's most likely salary classification within the lower and more densely packed region.

We then ran several optimisation techniques through the models we had trained on the data, ensuring that they were robust by also testing on unseen data and cross validating results rigorously to make sure we were not causing our model to be overly sensitive to known data, and ergo less effective against new unseen data. The resultant accuracy score we obtained suggested our model was 0.71, or 71%, suggesting it could correctly guess the salary classification (either above or below the sample median value) 7 times out of every 10.

##### Improvements

Of course a larger dataset and a longer timeframe to run the tools necessary to download the information from the web in the first place (as trying too much, too fast can result in being temporarily blocked) would be helpful, as the more data we have the more confidence we can have in our results.

Further dimensions being added to the job postings is another avenue where we could see some meaningful improvements, adding further keyword searches and phrases referring to the amount of prior experience needed, specific programming languages or other software being targeted in the job posting, for example.

The selection of models also performed different in different domains of the data, so given time to optimise a combination of different models, each being utilised within their best range of the data, we could end up with a more capable ensemble approach.

##### Conclusions

We ultimately arrived at a model which was notably more capable of identifying whether a role would be above or below the median salary level for our given sample than random approximation, although in its current form caution should be advised in being overly reliant on it as being incorrect 30% of the time can clearly cause significant issues depending on the context of the deployment of the model.

It should also be noted that if there were tolerances or biases that were of no consequence to a specific goal then further adjustments can be made to maximise various different criteria of the model. For example, should you be comfortable/should your business purpose have the necessary tolerance to accidentally classify below median salary roles as above median salary roles as a trade off for ensuring you classified as many truly above median salary roles correctly as practically possible, then adjustments can be made to such an effect (and vice versa, naturally).
